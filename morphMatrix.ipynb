```python
class Method:
  def __init__(self, power, estimate, overflow, setup):
    self.power = power
    self.estimate = estimate
    self.overflow = overflow
    self.setup = setup

  def toString(self):
    str = self.power + ", " + self.estimate + ", " + self.overflow + ", " + self.setup 
    return str
```

```python
powers = ["mechanical", "battery", "heat", "water", "sunlight"]
estimates = ["lever", "contactless water sensor", "beep for every 1ml", "tactile feedback", "computer vision"]
overflows = ["sound", "heat", "shock", "tactile feedback", "smell"]
setups = ["clip to cup", "wear on wrist", "wear on finger", "wrap over cup", "wear on head"]
```

```python
methods = []

for i in range(len(powers)):
  print("-------------")
  for estimate in estimates:
      m = Method(powers[i], estimate, overflows[i], setups[i])
      print(m.toString())
      methods.append(m)
  for overflow in overflows:
      m = Method(powers[i], estimates[i], overflow, setups[i])
      print(m.toString())
      methods.append(m)
  for setup in setups:
      m = Method(powers[i], estimates[i], overflows[i], setup)
      print(m.toString())
      methods.append(m)
```

```python
pStorage = []
eStorage = []
oStorage = []
sStorage = []

for method in methods:
  pStorage.append(method.power)
  eStorage.append(method.estimate)
  oStorage.append(method.overflow)
  sStorage.append(method.setup)
```

```python
import pandas as pd

d = {
    "Power Source": pStorage,
    "Estimate Water Level": eStorage,
    "Prevent water overflowing": oStorage,
    "Set Up": sStorage
}
df = pd.DataFrame(data=d)
df
```
